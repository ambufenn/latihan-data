#     Berapakah rata-rata dari data tersebut?
#     Berapakah banyak angka 20 muncul dalam angka tersebut?
#     Hapuslah data yang memiliki duplikasi, artinya data yang unik yah.


```python
nums = [20, 30, 10, 40, 20, 50, 60, 80, 0, 20, 30]

def calculate_average(nums):
    return round(sum(nums) / len(nums), 2)

def count_twenty(nums):
    return nums.count(20)

def remove_duplicates(nums):
    return list(set(nums))

print("The average is", calculate_average(nums))
print("The number 20 appears", count_twenty(nums), "times")
print("Unique numbers in the list are:", remove_duplicates(nums))
```

```python
nums = [20, 30, 10, 40, 20, 50, 60, 80, 0, 20, 30]

def calculate_average(nums):
    sum_nums = 0
    for num in nums:
        sum_nums += num
    return round(sum_nums / len(nums), 2)

def count_twenty(nums):
    count = 0
    for num in nums:
        if num == 20:
            count += 1
    return count

def remove_duplicates(nums):
    unique_nums = []
    for num in nums:
        if num not in unique_nums:
            unique_nums.append(num)
    return unique_nums

print("The average is", calculate_average(nums))
print("The number 20 appears", count_twenty(nums), "times")
print("Unique numbers in the list are:", remove_duplicates(nums))
```
