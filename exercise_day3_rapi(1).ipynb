```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the bars
nk = city[0:4]
ri = reject_item[0:4]
cmpl = complaint[0:4]

# Calculate the total value of each bar
totals = np.array(ri) + np.array(cmpl)

# Sort the bars in ascending order by their total value
sorted_indices = np.argsort(totals)
nk = np.array(nk)[sorted_indices]
ri = np.array(ri)[sorted_indices]
cmpl = np.array(cmpl)[sorted_indices]

# Define the colors for the bars
colors = ['#FF9999', '#00BFFF', '#C1FFC1', '#CAE1FF']

# Define the position of the bars on the x-axis
r = range(len(nk))
barWidth = 1

# Plot the bars in stack manner
plt.figure(figsize=(10,7))
plt.bar(r, ri, color=colors[0], edgecolor='white', width=barWidth, label="rejected item")
plt.bar(r, cmpl, bottom=ri, color=colors[1], edgecolor='white', width=barWidth, label='complaint')
plt.legend()

# Customize the X axis
plt.xticks(r, nk, fontweight='bold')

# Add a y-axis label
plt.ylabel("sales")

# Save the plot to a file and display it
plt.savefig("stacked1.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the bars
nk = city[0:4]
ri = reject_item[0:4]
cmpl = complaint[0:4]

# Calculate the total value of each bar
totals = np.array(ri) + np.array(cmpl)

# Sort the bars in ascending order by their total value
sorted_indices = np.argsort(totals)
nk = np.array(nk)[sorted_indices]
ri = np.array(ri)[sorted_indices]
cmpl = np.array(cmpl)[sorted_indices]

# Define the colors for the bars
colors = ['#FF9999', '#00BFFF', '#C1FFC1', '#CAE1FF']

# Define the position of the bars on the x-axis
r = np.arange(len(nk)) * 2
barWidth = 0.8

# Plot the bars in stack manner
plt.figure(figsize=(10,7))
plt.bar(r, ri, color=colors[0], edgecolor='white', width=barWidth, label="rejected item")
plt.bar(r, cmpl, bottom=ri, color=colors[1], edgecolor='white', width=barWidth, label='complaint')
plt.legend()

# Customize the X axis
plt.xticks(r, nk, fontweight='bold')

# Add a y-axis label
plt.ylabel("sales")

# Adjust the x-axis limits and add some padding
plt.xlim([-1, len(nk) * 2])
plt.xticks(np.arange(len(nk)) * 2, nk, fontweight='bold')

# Save the plot to a file and display it
plt.savefig("stacked2.png")
plt.show()
```
