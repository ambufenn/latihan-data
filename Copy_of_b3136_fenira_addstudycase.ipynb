Pada saat ini, AAPL memiliki dominanansi dalam kapitalisasi pasar. Mari plot kapitalisasi pasar untuk 10 stocks teratas sebagai plot bar untuk memvisualisasikan ini dengan lebih baik.

````python
# get data new, with company (unnamed) and marketcap

new_data_2 = df[['Unnamed: 0','marketCap']]

new_data_2 = new_data_2.rename(columns={'Unnamed: 0': 'Capital'})

new_data_2 = new_data_2.sort_values(by='marketCap', ascending=True)

nan_values = new_data_2.isna()

new_data_2 = new_data_2.dropna()

new_data_2.head(5)
````

````python
count_row_marketCap = new_data_2['marketCap'].count()
count_row_marketCap

# ok cocok ya
````

````python
top_10 = new_data_2.nlargest(10, 'marketCap')
top_10.head(4)
````

````python
# Calculate total market cap
total_cap = new_data_2['marketCap'].sum() # here the sum = 17183740754108.0
 
# Add new column for % of total cap
top_10['percentage_marketCap'] = 0

top_10.head(4)
# Calculate % of total cap for each company

top_10['percentage_marketCap'] = (top_10['marketCap'] / total_cap) * 100
top_10.head(4)

top_10['percentage_marketCap'] = top_10['percentage_marketCap'].round(1)

top_10.head(4)

# # Sort by descending % of total cap
top_10 = top_10.sort_values('percentage_marketCap', ascending=False)

# Print top 10 companies by % of total cap
top_10 = top_10.reset_index(drop=True)
top_10.head(10)
````

````python
import matplotlib.pyplot as plt

a = top_10
	
ax = a.plot.bar(x="Capital", y="marketCap", rot=0)

#bisa grafixnya, tinggal di css di hias , okey
````

````python
#coba perindah grafix

import matplotlib.pyplot as plt
import pandas as pd

ax = a.plot.bar(x="Capital", y="marketCap", rot=0, legend=False, width=0.7, color=["#4285F4"])

# Remove top and right spines
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Set y-axis label
ax.set_ylabel("Percentage", fontsize=12, fontweight="bold")

# Set x-tick label font size
ax.tick_params(axis="x", labelsize=10)

# Set y-tick label font size
ax.tick_params(axis="y", labelsize=10)

# Set x-axis label to None
ax.set_xlabel(None)

# Set title
ax.set_title("Top 10 Stocks by Percentage", fontsize=14, fontweight="bold")

# Add value labels to the bars
for i, v in enumerate(a["percentage_marketCap"]):
    ax.text(i, v+0.1, str(v)+"%", ha="center", fontsize=10)

plt.show()

#better dr yg awal, tp masi ada yg beda dr contoh answer. nanti di hias lagi ya
````

**Membuat plot lebih mudah dibaca dan lebih informatif**

Meskipun plot di atas cukup informatif, namun dapat kamu diperbaiki. Ada beberapa stocks terlalu besar, dan stocks lainnya sulit dibedakan karena hal ini. Alih-alih persentase, mari gunakan skala log10 dari market kapitalisasi. 

````python
import matplotlib.pyplot as plt

# x-axis: market capitalization
x = a['Capital']

# y-axis: market cap value in USD divided by 10 billion
y = a['marketCap'] / 1e10

# create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x, y)

# set x-axis label and tick labels
ax.set_xlabel('capital')
ax.set_xticklabels(['${:,.0f}'.format(xtick / 1e9) + 'B' for xtick in ax.get_xticks()])
ax.set_xlim([0, ax.get_xlim()[1]])  # adjust x-axis limit to start at 0

# set y-axis label and tick labels
ax.set_ylabel('Market Cap (USD in billions)')
ax.set_yticklabels(['${:,.0f}'.format(ytick * 10) + 'B' for ytick in ax.get_yticks()])

# set plot title
ax.set_title('capital vs. Market Cap Value in USD')

# display the plot
plt.show()

#untuk yg ini masi rancu marketcap usd nya itu value yg mana, kita coba baca ulang ya arahnya
````

````python
# Definisikan fungsi dengan 2 parameter (series untuk plot dan title)
def top10_subplot(volatility_series, title):
    # Buat subplot untuk dua sisi 
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
    
    # Visualisasi untuk 10 terbawah dengan pandas barchart
    ax = volatility_series[:10].plot(kind='barh', color='red', ax=axes[0])
    
    # Buat judul disini
    ax.set_title('Losers')
    
    # Namakan ylabel '% change'
    ax.set_ylabel('% change')
    
    # Sama seperti diatas untuk mengambil 10 teratas
    ax = volatility_series[-10:].plot(kind='barh', color='green', ax=axes[1])
    
    # menghasilkan graphnya
    return fig, ax

DTITLE = "50 days top losers and winners"

# Panggil fungsi untuk 50days dengan judul DTITLE 
# fig, ax = top10_subplot(volatility[fiftyDayAverageChangePercent], DTITLE)

fig, ax = top10_subplot(volatility['fiftyDayAverageChangePercent'], DTITLE)
````
