```python
produk_dict = {'Pensil': [25, 2000, 0], 'Buku': [30, 0, 60000], 'Bolpoin': [10, 0, 20000], 'Kertas': [10, 500, 5000], 'Tas': [5, 50000, 0], 'Penghapus': [50, 0, 24000], 'Busur': [15, 0, 30000], 'Penggaris': [15, 1000, 0], 'Spidol': [50, 0, 100000], 'Meja Tulis': [4, 0, 600000], 'Papan Tulis': [2, 0, 600000], 'Lemari': [5, 0, 800000], 'Kursi': [5, 85000, 0], 'Sepatu': [20, 200000, 0], 'Kaos Kaki': [45, 10000, 0]}

total_penjualan_kurang_50k = 0

for produk, detail in produk_dict.items():
    penjualan = detail[2]
    if penjualan < 50000:
        total_penjualan_kurang_50k += penjualan

print(total_penjualan_kurang_50k)

pendapatan = total_penjualan_kurang_50k * 0.2

print("Pendapatan dari penjualan produk dengan nilai penjualan kurang dari 50000 adalah: Rp", pendapatan)
```
